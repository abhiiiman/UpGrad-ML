## Adding `.` instead of `,` in the required columns

```python
column_fix = ['CO(GT)', 'C6H6(GT)','T', 'RH', 'AH']
air_df[column_fix] = air_df[column_fix].applymap(lambda x: str(x).replace(',', '.'))
```

```python
air_df
```

## Converting the cleaned columns to float

```python
air_df[column_fix] = air_df[column_fix].astype("float64")
```

```python
air_df.info()
```

## Checking the datatype of `Date` and converting it to `datetime`

```python
air_df['Date']
```

```python
air_df['Date'] = pd.to_datetime(air_df['Date'], format='%d/%m/%Y')
air_df["Date"].info()
air_df["Date"]
```

## Converting `Time` to date time

```python
air_df['Time'] = pd.to_datetime(air_df['Time'], format='%H.%M.%S').dt.time
air_df["Time"].info()
```

## Concatenating both `Date` and `Time` to `DateTime`

```python
air_df['Date'] = air_df['Date'] + pd.to_timedelta(air_df["Time"].astype(str))
```

```python
air_df
```

## Dropping the Time column

```python
air_df = air_df.drop(columns = ['Time'])
air_df
```

## Renaming the `Date` to `DateTime`

```python
air_df.rename(columns = {"Date": "DateTime"}, inplace = True)
air_df
```

```python
air_df.info()
```
