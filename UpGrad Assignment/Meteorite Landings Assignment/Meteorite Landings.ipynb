## Latitude ranges from -90 to 90 and Longitude ranges from -180 to 180

```python
df = df[(df['reclat'] >= -90) & (df['reclat'] <= 90) & (df['reclong'] >= -180) & (df['reclong'] <= 180)]
df
```

## Converting the categorical features

```python
df['fall'] = df['fall'].astype('category')
df['fall'].info()
```

```python
df['nametype'] = df['nametype'].astype('category')
```

## Strip leading and trailing spaces from object columns

```python
df['name'] = df['name'].str.strip()
df['GeoLocation'] = df['GeoLocation'].str.strip()
```

## Detecting and Removing the Outliers

```python
Q1 = df['mass'].quantile(0.25)
Q3 = df['mass'].quantile(0.75)
IQR = Q3 - Q1
df = df[(df['mass'] >= Q1 - 1.5 * IQR) & (df['mass'] <= Q3 + 1.5 * IQR)]
```
